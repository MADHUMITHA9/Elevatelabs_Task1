```python
import pandas as pd
df = pd.read_csv("C:/Users/user/Desktop/netflix_titles.csv")
print("----- Dataset Preview -----")
print(df.head())
print("\n----- Basic Info -----")
print(df.info())
print("\n----- Missing Values -----")
print(df.isnull().sum())
df = df.assign(
    director=df['director'].fillna("Unknown"),
    cast=df['cast'].fillna("Not Available"),
    country=df['country'].fillna("Unknown"),
    rating=df['rating'].fillna("Not Rated"),
    date_added=df['date_added'].fillna("Unknown")
)
duplicates_count = df.duplicated().sum()
print(f"\nDuplicates Found: {duplicates_count}")
df = df.drop_duplicates()
df['type'] = df['type'].str.lower()
df['country'] = df['country'].str.lower()
df['rating'] = df['rating'].str.upper()
df['listed_in'] = df['listed_in'].str.lower()
df['title'] = df['title'].str.title()
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
df.columns = df.columns.str.lower().str.replace(' ', '_')
print("\n----- Data Types After Cleaning -----")
print(df.dtypes)
print("\n----- Summary Statistics -----")
print(df.describe(include='all'))
df.to_csv("C:/Users/user/Desktop/cleaned_netflix_dataset.csv", index=False)
print("\nâœ… Cleaned dataset saved successfully as 'cleaned_netflix_dataset.csv'")
print("\nðŸ“‹ Summary of Changes:")
print("""
1. Removed duplicate rows.
2. Filled missing values for director, cast, country, rating, and date_added.
3. Converted 'date_added' column to proper datetime format.
4. Standardized text values (lowercase / title case).
5. Renamed columns (lowercase, underscores instead of spaces).
6. Verified and fixed data types.
7. Exported final cleaned dataset.
""")
```
